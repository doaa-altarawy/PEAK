all_genes = np.unique(true.as_matrix())
all_genes = pd.Series(all_genes).replace(mapping_basic).unique()

mapping = pd.DataFrame(columns=['true_gene',  'gene_id', 'short_name', 'AGI'])
mapping.true_gene = all_genes

# Genes with parenthesis 
# print(mapping[mapping.true_gene.str.contains("\(")])
mapping.set_index('true_gene', inplace=True)

missing = []
for gene in mapping.index:
    # search in short names
    found = data[data.AGI.apply(lambda x: True if str(x).find(gene)>=0 else False)]
    if (found.shape[0]>0):
        mapping.loc[gene] = found[['gene_id', 'gene_short_name', 'AGI']].values
    else:
        found = data[data.gene_short_name.apply(lambda x: True if str(x).find(gene)>=0 else False)]
        if (found.shape[0]==1):
            mapping.loc[gene] = found[['gene_id', 'gene_short_name', 'AGI']].values
        elif (found.shape[0]>1):
            exact = data[data.gene_short_name==gene]
            if (exact.shape[0]>0):
                mapping.loc[gene] = exact[['gene_id', 'gene_short_name', 'AGI']].values
            else:
                print(gene, found.shape[0])
        else:
            print('Missing Gene: ', gene)
            missing.append(gene)
    

mapping.to_csv(final_mapping)
